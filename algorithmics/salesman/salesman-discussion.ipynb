```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
from salesman import TSPSolver 
```

```python
# Function to run the TSP solver and collect data
def run_solver(file_path, neighborhood_method):
    tsp_solver = TSPSolver(file_path, neighborhood_method)
    start_time = time.time()
    # Unpacking only the necessary data from the method
    _, best_distance, _, _ = tsp_solver.steepest_hill_climbing_with_restarts(max_restarts=5, max_iterations=100)
    end_time = time.time()
    computation_time = end_time - start_time
    return best_distance, computation_time
```

```python
# File to test
file_path = "tsp25.txt"  # Change this to your actual file

# Run the solver for each neighborhood method
results = []
for method in ["2-opt", "2-swap"]:
    distance, comp_time = run_solver(file_path, method)
    results.append({
        'neighborhood_method': method,
        'final_distance': distance,
        'computation_time': comp_time
    })

# Convert results to a DataFrame
results_df = pd.DataFrame(results)
```

```python
# Plotting
plt.figure(figsize=(12, 5))

# Distance comparison
plt.subplot(1, 2, 1)
plt.bar(results_df['neighborhood_method'], results_df['final_distance'], color='skyblue')
plt.xlabel('Neighborhood Method')
plt.ylabel('Final Distance')
plt.title('Final Distance by Neighborhood Method')

# Time comparison
plt.subplot(1, 2, 2)
plt.bar(results_df['neighborhood_method'], results_df['computation_time'], color='lightgreen')
plt.xlabel('Neighborhood Method')
plt.ylabel('Computation Time (seconds)')
plt.title('Computation Time by Neighborhood Method')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time

# Define the file names and neighborhood methods
file_names = ["tsp5.txt","tsp25.txt","tsp50.txt","tsp101.txt" ]
neighborhood_methods = ["2-opt", "2-swap"]

# Number of tries for averaging
num_tries = 4

# Initialize arrays to store computation times
computation_times = np.zeros((len(file_names), len(neighborhood_methods), num_tries))

# Function to run steepest_hill_climbing and measure computation time
def run_and_measure_computation_time(file_path, neighborhood_method, max_iterations):
    tsp_solver = TSPSolver(file_path, neighborhood_method)
    start_time = time.time()
    tsp_solver.steepest_hill_climbing(max_iterations)
    end_time = time.time()
    return end_time - start_time

# Run the computation for each combination of file and method multiple times
for file_index, file_name in enumerate(file_names):
    for method_index, method in enumerate(neighborhood_methods):
        for try_index in range(num_tries):
            # Calculate computation time
            time_taken = run_and_measure_computation_time(file_name, method, max_iterations=1000)
            computation_times[file_index, method_index, try_index] = time_taken

# Calculate average computation times
average_computation_times = np.mean(computation_times, axis=2)

# Plot the average computation times
plt.figure(figsize=(10, 6))

for method_index, method in enumerate(neighborhood_methods):
    plt.plot(file_names, average_computation_times[:, method_index], 'o-', label=method)

plt.xlabel('File Name')
plt.ylabel('Average Computation Time (Seconds)')
plt.title('Average Computation Time for Different Files and Neighborhood Methods (10 Tries)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import time

# Define the file names and neighborhood methods
file_names = ["tsp5.txt", "tsp25.txt"]
neighborhood_methods = ["2-opt", "2-swap"]

# Number of tries for averaging
num_tries = 2

# Initialize arrays to store computation times
computation_times = np.zeros((len(file_names), len(neighborhood_methods), num_tries))

# Function to run tabu_search without tabu list and measure computation time
def run_and_measure_computation_time_without_tabu(file_path, neighborhood_method, max_moves):
    tsp_solver = TSPSolver(file_path, neighborhood_method)
    start_time = time.time()
    tsp_solver.tabu_search(max_moves, tabu_size=1)  # Use tabu_size=0 to disable tabu list
    end_time = time.time()
    return end_time - start_time

# Run the computation for each combination of file and method multiple times
for file_index, file_name in enumerate(file_names):
    for method_index, method in enumerate(neighborhood_methods):
        for try_index in range(num_tries):
            # Calculate computation time
            time_taken = run_and_measure_computation_time_without_tabu(file_name, method, max_moves=1000)
            computation_times[file_index, method_index, try_index] = time_taken

# Calculate average computation times
average_computation_times = np.mean(computation_times, axis=2)

# Plot the average computation times
plt.figure(figsize=(10, 6))

for method_index, method in enumerate(neighborhood_methods):
    plt.plot(file_names, average_computation_times[:, method_index], 'o-', label=method)

plt.xlabel('File Name')
plt.ylabel('Average Computation Time (Seconds)')
plt.title('Average Computation Time for Tabu Search (Without Tabu List) for Different Files and Neighborhood Methods (10 Tries)')
plt.legend()
plt.grid(True)
plt.show()
```
